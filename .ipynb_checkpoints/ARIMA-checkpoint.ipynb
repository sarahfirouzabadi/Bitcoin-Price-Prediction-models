```python
X_new=pd.DataFrame(X_Price)
```

```python
X_new
```

```python
X_new['difference']=np.log(X_new).diff()
```

```python
X_new.isna().sum()
```

```python
X_new.dropna(inplace=True)
```

```python
X_new=X_new['difference']
```

```python
X_new
```

```python
X_new.shape
```

```python
X_train=X_new[:1600]
X_test=X_new[1600:]
```

```python
model =sm.tsa.arima.ARIMA(X_train)
model_fit = model.fit()
# print(model_fit.summary())
```

```python
yhat=model_fit.predict(start=1600,end=2008)
```

```python
yhat.shape
```

```python
Y_real_test=X_test
```

```python
Y_real_test.shape
```

```python
np.sqrt(mean_squared_error(Y_real_test,yhat))
```

```python
mean_squared_error(Y_real_test,yhat)
```

```python
mean_absolute_error(Y_real_test,yhat)
```
